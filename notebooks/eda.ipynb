### Articles publication date analysis

```python
# Count articles by date
date_counts = analyst_ratings['date'].value_counts().sort_index()

# Plot the publication frequency over time
plt.figure(figsize=(12, 6))
date_counts.plot(kind='line')
plt.title('Article Publication Frequency Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.show()
```

### Articles per publishers

```python
top_publishers = publisher_counts.head(10)  
plt.figure(figsize=(10, 6))
top_publishers.plot(kind='bar', color='skyblue')
plt.title('Top 10 Publishers by Number of Articles')
plt.xlabel('Publisher')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.show()
```

## Sentiment Analysis & Topic Modeling

### Sentiment Analysis

```python
def classify_sentiment(compound_score):
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'
# Apply classification
analyst_ratings['sentiment_class'] = analyst_ratings['sentiment'].apply(classify_sentiment)

# Display the first few rows
print(analyst_ratings[['headline', 'sentiment', 'sentiment_class']].head())
```

## 3) Time Series Analysis

```python
# Convert 'date' column to datetime, letting pandas infer the format
analyst_ratings['date'] = pd.to_datetime(analyst_ratings['date'], errors='coerce')

# Check for any failed conversions (NaT indicates a failed conversion)
print(analyst_ratings[analyst_ratings['date'].isna()])

# Now extract the hour
analyst_ratings['hour'] = analyst_ratings['date'].dt.hour

# Count the occurrences of each hour
hour_counts = analyst_ratings['hour'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(10, 6))
hour_counts.plot(kind='bar')
plt.title('Article Publication Frequency by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Articles')
plt.xticks(rotation=0)
plt.show()
```
